# Stationary check

```python
# Determining rolling statistics

rollmean=indexedataset.rolling(window=12).mean()
rollstd=indexedataset.rolling(window=12).std()
print(rollmean,rollstd)
```

```python
# plot rolling statistics

orig=plt.plot(indexedataset,color="blue",label="Original")
mean=plt.plot(rollmean,color="red",label="rolling mean")
std=plt.plot(rollstd,color="black",label="rolling std")
plt.legend(loc="best")
plt.title("Rolling mean & standard deviation")
plt.show(block=False)
```

```python
# here it looks like there is no trends here
```

```python
# Perform Dicky fuller test

from statsmodels.tsa.stattools import adfuller
print("Dicky fuller test results:")
df_test=adfuller(indexedataset["Sales"],autolag="AIC")
dfoutput=pd.Series(df_test[0:4],index=["Test Statistic","P-Value","Lags used","No.of Observations Used"])
for key,value in df_test[4].items():
    dfoutput["Critical value (%s)"%key]=value
print(dfoutput)
```

```python
# H0 - it is not stationary
# H1 - it is stationary
# here P-value=0.0 which is less than 0.05 , that means it rejects the null hyppthesis , that means our data is stationary
```

```python
#estimating trend
indexedataset_log=np.log(indexedataset)
plt.figure(figsize=(10,6))
plt.plot(indexedataset_log)
```

# Moving Averages

```python
moving_avg=indexedataset_log.rolling(window=12).mean()
moving_std=indexedataset_log.rolling(window=12).std()
plt.figure(figsize=(10,6))
plt.plot(indexedataset_log)
plt.plot(moving_avg,color="red")
plt.plot(moving_std,color="black")
```

```python
# here it is better than previous one
```

```python
indexedataset_log.isnull().sum()
```

```python
exponentDecayWeighted_avg=indexedataset_log.ewm(halflife=12,min_periods=0,adjust=True).mean()
plt.figure(figsize=(10,6))
plt.plot(indexedataset_log)
plt.plot(exponentDecayWeighted_avg,color="red")
```

```python
#ACF and PACF plots to find values of p and q in ARIMA 

from statsmodels.tsa.stattools import acf,pacf
lag_acf=acf(indexedataset_log,nlags=10)
lag_pacf=pacf(indexedataset_log,nlags=10)

#plot ACF
plt.figure(figsize=(10,6))
plt.plot(lag_acf)
plt.axhline(y=0,linestyle="--",color="gray")
plt.axhline(y=-1.96/np.sqrt(len(indexedataset_log)),linestyle="--",color="gray")
plt.axhline(y=1.96/np.sqrt(len(indexedataset_log)),linestyle="--",color="gray")
plt.title("Auto Correlation Plot")
plt.show()

#plot PACF
plt.figure(figsize=(10,6))
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle="--",color="gray")
plt.axhline(y=-1.96/np.sqrt(len(indexedataset_log)),linestyle="--",color="gray")
plt.axhline(y=1.96/np.sqrt(len(indexedataset_log)),linestyle="--",color="gray")
plt.title("Partial Auto Correlation Plot")
plt.show()
```

```python
# here both lines meet at 0 at point 3,3 respectively, hence we take p=3,q=3 in ARIMA model
```

# Model Selection and Forecasting:

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Auto Regression Model (AR)

ar_model=ARIMA(indexedataset_log,order=(0,0,3))
results_ar=ar_model.fit()
plt.figure(figsize=(10,6))
plt.plot(indexedataset_log)
plt.plot(results_ar.fittedvalues,color="red")
plt.title("RSS:%.4f"%sum(results_ar.fittedvalues-indexedataset_log["Sales"])**2)
print("---------------AR Model-----------------")
```

```python
# Moving Average Model (AR)

ma_model=ARIMA(indexedataset_log,order=(3,0,0))
results_ma=ma_model.fit()
plt.figure(figsize=(10,6))
plt.plot(indexedataset_log)
plt.plot(results_ma.fittedvalues,color="red")
plt.title("RSS:%.4f"%sum(results_ma.fittedvalues-indexedataset_log["Sales"])**2)
print("---------------MA Model-----------------")
```

```python
# Autoregressive integrated moving average model
arima_model=ARIMA(indexedataset_log,order=(3,0,3))
results_arima=arima_model.fit()
plt.figure(figsize=(10,6))
plt.plot(indexedataset_log)
plt.plot(results_arima.fittedvalues,color="red")
plt.title("RSS:%.4f"%sum(results_arima.fittedvalues-indexedataset_log["Sales"])**2)
print("---------------ARIMA Model-----------------")
```

```python
# here RSS value of ARIMA is very less , so let's consider the model as ARIMA
```

```python
pred_arima=pd.Series(results_arima.fittedvalues,copy=True)
pred_arima
```

```python
# convert to cumulative sum
pred_arima_cumsum=pred_arima.cumsum()
pred_arima_cumsum
```
